```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('figure',figsize = (15,8))
```

```python
dataset = pd.read_csv('destino_plastico.csv')
```

```python
dataset
```

```python
dataset_fix = dataset[['Entity','Year','Estimated historic plastic fate']]
dataset_fix
```

```python
destino = dataset_fix['Entity']
```

```python
destino.drop_duplicates()
```

```python
dataset_fix.query('Entity == "Discarded"').plot.line(x = 'Year',y ='Estimated historic plastic fate',legend = False, figsize =(10,8), xlabel = 'Anos', ylabel = '% de todo o plastico',title = 'Plástico Descartado Global')
```

```python
dataset_fix.query('Entity == "Incinerated"').plot.line(x = 'Year',y ='Estimated historic plastic fate',legend = False, figsize =(10,8), xlabel = 'Anos', ylabel = '% de todo o plastico',title = 'Plástico Incinerado Global')
```

```python
dataset_fix.query('Entity == "Recycled"').plot.line(x = 'Year',y ='Estimated historic plastic fate',legend = False, figsize =(10,8), xlabel = 'Anos', ylabel = '% de todo o plastico',title = 'Plástico Reciclado Global')
```

```python
area = plt.figure()
```

```python
descartado = area.add_subplot(2,2,1)
incinerado = area.add_subplot(2,2,2)
reciclado = area.add_subplot(2,2,3)
```

```python
selecao = dataset_fix.query('Entity == "Discarded"')
descartado.plot(selecao['Year'],selecao['Estimated historic plastic fate'])
descartado.set_title('Porcentagem de Plastico Descartado Global')
descartado.set_xlabel('Anos')
descartado.set_ylabel('% do Plastico')
```

```python
selecao = dataset_fix.query('Entity == "Incinerated"')
incinerado.plot(selecao['Year'],selecao['Estimated historic plastic fate'])
incinerado.set_title('Porcentagem de Plastico Incinerado Global')
incinerado.set_xlabel('Anos')
incinerado.set_ylabel('% do Plastico')
```

```python
selecao = dataset_fix.query('Entity == "Recycled"')
reciclado.plot(selecao['Year'],selecao['Estimated historic plastic fate'])
reciclado.set_title('Porcentagem de Plastico Reciclado Global')
reciclado.set_xlabel('Anos')
reciclado.set_ylabel('% do Plastico')
```
